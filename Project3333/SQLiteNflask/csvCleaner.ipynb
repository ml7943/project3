```python
# Setup Dependencies
import pandas as pd
```

```python
# Read csv file and create a dataframe
nyc_crime_df = pd.read_csv("Resources/crime.csv")    
```

```python
# Preview the DataFrame
nyc_crime_df.head()
```

```python
# Drop the columns that are not needed  
columns_to_drop = ['KY_CD', 'ADDR_PCT_CD', 'PD_DESC', 'LOC_OF_OCCUR_DESC', 'PREM_TYP_DESC', 'SUSP_AGE_GROUP', 'SUSP_RACE','SUSP_SEX', 'VIC_AGE_GROUP', 'VIC_RACE', 'VIC_SEX', 'Zip Codes', 'City Council Districts']
nyc_crime_df = nyc_crime_df.drop(columns=columns_to_drop)   
nyc_crime_df.head() 
```

```python
# Rename columns})
nyc_crime_df.index = ["" for _ in range(len(nyc_crime_df))]
nyc_crime_df = nyc_crime_df.rename(columns={'CMPLNT_NUM':'Complaint No.', 'CMPLNT_FR_DT':'Complaint Date', 'OFNS_DESC': 'Offense',	'LAW_CAT_CD': 'Law Cat', 'BORO_NM': 'Borough'})
nyc_crime_df.head()
```

```python
# Check the data types
nyc_crime_df.dtypes
```

```python
# Export cleaned csv file to Resources folder

pd.DataFrame.to_csv(nyc_crime_df, 'shrunk_crime.csv', index=False)
```
